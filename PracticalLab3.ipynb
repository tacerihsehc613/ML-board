```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np

price = datasets.load_boston()
X_train, X_test, y_train, y_test=train_test_split(price.data, price.target)

lr=LinearRegression()
lr.fit(X_train,y_train)

predicted=lr.predict(X_test)
plt.scatter(y_test,predicted)
R2=np.round(r2_score(y_test, predicted),2)
plt.title('Housing Prices, split (80/20) - R2: {R2}'.format(R2=R2))
# CODE PROVIDED TO COMPARE YOUR RESULTS AGAINST THE ACTUAL LABELS
xlim = [0, 55]
ylim = [0, 55]
plt.plot(xlim, ylim, '--k', label=f"True labels")
plt.xlabel("True price ($1K)")
plt.ylabel("Predicted price ($1K)")
plt.legend()
plt.show()
```

```python
iris_target_names = ["Setosa", "Versicolor", "Virginica"]

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_boston

iris=datasets.load_iris()
irisDF=pd.DataFrame(iris.data,columns=iris['feature_names'])

irisDF['type']=iris.target
```

```python
import seaborn as sns
sns.set_style("whitegrid");
sns.pairplot(irisDF,kind="scatter",hue="type",size=3);
plt.show()

"""
Through bivariate analysis of each feature, petal_length and petal_width worked best for classifying flowers. 
This is because, in the relationship between petals_length and petals_width, 
there exists the least overlap between different flower types.
"""
```

```python
corr=irisDF.drop(['type'],axis=1).corr()
sns.heatmap(corr,annot=True,fmt='.1f',cbar=True)
plt.show()

"""
YOUR BRIEF ANSWER TO WHAT A CORRELATION PLOT MEANS HERE:
A correlation plot measures the strength between two features. 
The further away the correlation coefficient from 0, the stronger the relationship between the two variables.
The highest correlation between petals_length and petals_width means they are most linearly dependent
so they also come across as the best features for separating the data into labels.
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = irisDF.iloc[:, 0:-1]
y=irisDF.iloc[:, -1]


X_train, X_test, Y_train, Y_test=train_test_split(X,y,test_size = 0.2, random_state=101)
lr=LogisticRegression()
lr.fit(X_train,Y_train)
pred=lr.predict(X_test)
acc=lr.score(X_test,Y_test)

assert acc >= 0.95, "You haven’t reached a high enough accuracy."
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
import seaborn as sns

cf_matrix = metrics.confusion_matrix(pred, Y_test)
print("Confusion Matrix: ")
print(cf_matrix)
print()

plt.imshow(cf_matrix,
           interpolation='nearest', cmap=plt.cm.binary)
plt.grid(False)
plt.colorbar()
plt.xlabel("predicted label")
plt.ylabel("true label");

sns.heatmap(cf_matrix, annot=True, cmap='Blues')
```
